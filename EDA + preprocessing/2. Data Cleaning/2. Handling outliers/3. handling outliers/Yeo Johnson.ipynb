```python
import pandas as pd
from sklearn.preprocessing import PowerTransformer

# Sample DataFrame
data = {'feature_col': [-100, -20, -15, -13, -12, -11, -10, -9, -5, 0, 5, 10, 11, 12, 13, 15, 20, 150]}
train = pd.DataFrame(data)

# Apply Yeo-Johnson transformation
pt = PowerTransformer(method='yeo-johnson', standardize=False)
train['feature_col_transformed'] = pt.fit_transform(train[['feature_col']])

# Plotting to visualize the effect (optional but recommended)
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(train['feature_col'], kde=True, ax=ax[0])
ax[0].set_title("Original Distribution")
sns.histplot(train['feature_col_transformed'], kde=True, ax=ax[1])
ax[1].set_title("Yeo-Johnson Transformed Distribution")
plt.show()

print(train[['feature_col', 'feature_col_transformed']])
```
