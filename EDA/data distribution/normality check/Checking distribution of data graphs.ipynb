```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.api.types import is_numeric_dtype
import scipy.stats as stats
```

```python
train = pd.read_csv(r"C:\Users\basde\Downloads\playground-series-s3e24\train.csv")
```

```python
numeric_cols = train.select_dtypes(include=np.number).columns
```

```python
for col in numeric_cols:
    plt.figure(figsize=(8, 5))
    
    # 1. Plot the histogram without the KDE line
    ax = sns.histplot(train[col], kde=False)
    
    # 2. Add the KDE line on the same plot, using the second y-axis
    ax2 = ax.twinx()
    sns.kdeplot(train[col], bw_adjust=5, ax=ax2, color='r') # Use ax2
    
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
for column in train.columns:
    # Check if the column contains numerical data
    if is_numeric_dtype(train[column]):
        # Create a new figure and axes for each plot
        plt.figure(figsize=(6, 4))
            
        # Create the boxplot using seaborn
        sns.boxplot(y=train[column])
            
        # Set the title for the plot
        plt.title(f'Boxplot of {column}', fontsize=14)
            
        # Display the plot
        plt.show()
```

```python
print("Generating Q-Q plots for each numeric column...")
for col in numeric_cols:
    # Create a figure and axis for the plot
    plt.figure(figsize=(8, 6))
    
    # Generate the Q-Q plot using scipy.stats.probplot
    # 'probplot' generates a probability plot, which for a normal distribution
    # is a Q-Q plot. It plots the data against the quantiles of a
    # specified theoretical distribution (by default, the normal distribution).
    stats.probplot(train[col], dist="norm", plot=plt)
    
    # --- Add titles and labels for clarity ---
    plt.title(f'Q-Q Plot for {col}', fontsize=16)
    plt.xlabel('Theoretical Quantiles (Normal Distribution)', fontsize=12)
    plt.ylabel('Sample Quantiles', fontsize=12)
    
    # Display the plot
    plt.show()

print("\nFinished generating plots.")
```
